# NLP Library Description

1. NLTK : This is one of the most usable and mother of all NLP libraries.
2. spaCy : This is completely optimized and highly accurate library widely used in deep learning
3. Stanford CoreNLP Python : For client-server based architecture this is a good library in NLTK. 
4. TextBlob : This is an NLP library which works in Pyhton2 and python3. This is used for processing textual data and              provide mainly all type of operation in the form of API.
5. Gensim : Genism is a robust open source NLP library support in python. This library is highly efficient and                    scalable.
6. Pattern : It is a light-weighted NLP module. This is generally used in Web-mining, crawling or such type of                     spidering task. 
7. Polyglot : For massive multilingual applications, Polyglot is best suitable NLP library. Feature extraction in the             way on Identity and Entity.
8. PyNLPl : PyNLPI also was known as 'Pineapple' and supports Python. It provides a parser for many data format like              FoLiA/Giza/Moses/ARPA/Timbl/CQL.
9. Vocabulary : This library is best to get Semantic type information from the given text.

# Explanation of the above code
1. We will import the SpeechRecognition library as sr.
2. The primary purpose of a sr.Recognizer instance is, of course, to recognize speech. Each instance comes with a        variety of settings and functionality for recognizing speech from an audio source.
3. With sr.Microphone as source we are giving access to Microphone to record the audio.
4. To handle ambient noise, you’ll need to use the adjust_for_ambient_noise() method of the Recognizer class to make    sense of the noisy audio file. Since input from a microphone is far less predictable than input from an audio        file, it is a good idea to do this anytime we listen for microphone input.
5. Just like the AudioFile class, Microphone is a context manager. We can capture input from the microphone using        the listen() method of the Recognizer class inside of the with block. This method takes an audio source as its        first argument and records input from the source until silence is detected.
6. Due to some instances if the audio is not clear, to check this we used try and except. If the audio is not            recognised we give the error message.
7. We try to recognise the speech from the audio with the help of r.recognise_google().


# Video to Text Conversion 

NLP (Natural Language Processing) is the application of computational techniques to the analysis and synthesis of natural language and speech. We can convert Text to Speech and Speech to text.


In our project, we have to convert the audio that is spoken in a video to text. For this, we must first convert the video file (.mp4) to a suitable audio format such as .wav format and then apply the code to convert the speech to text using speech_recognition module.

```python
import os
#it uses operating system
```

```python
import csv
#comma seperated value file
```

```python
cmdline = "ffmpeg -i C:/Users/Yash/testvideo.mp4 -ab 160k -ac 2 -ar 44100 -vn audio1.wav"
os.system(cmdline)
```

```python
subprocess.call(cmdline, shell=True)
# here the subprocess is making a call to cmdline 
r = sr.Recognizer()

#these is source recognizer
inp = sr.AudioFile('audio1.wav')

with inp as source:
    audio = r.record(source)
type(audio)
sa=r.recognize_google(audio)

print(sa)

#it will print the audio file .wav using google recognizer

with open('Book1.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=' ',quotechar='|', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(sa)
    
#we will use csv file to get the final text.
```
